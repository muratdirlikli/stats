## Q1 & Q2 & Q3 & IQR

```python
data = [32, 36, 46, 47, 56, 69, 75, 79, 79, 80, 88, 89]
```

```python
Q1=np.quantile(data,0.25)
Q1
```

```python
np.percentile(data,25)
```

```python
Q2=np.quantile(data,0.50)
Q2
```

```python
np.median(data)
```

```python
np.percentile(a=data,q=50)
```

```python
Q3=np.percentile(data,75)
Q3
```

```python
IQR=Q3-Q1
```

```python
print(f" IQR: {IQR}")
```
